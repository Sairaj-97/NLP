```python
import spacy

nlp = spacy.blank("en")

doc = nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token)
```

We get above error because we have a blank pipeline as shown below. Pipeline is something that starts with a Tokenizer component in a dotted rectange below. You can see there is nothing there hence the blank pipeline

```python
nlp = spacy.load("en_core_web_sm")
nlp.pipe_names
```

```python
doc = nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token, " | ", token.pos_ ," | ", token.lemma_)
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python

# Sample text
doc2 = nlp("Apple is looking at buying U.K. startup for $1 billion in 2023.")

# View each token with pipeline attributes
for token in doc2:
    print(f"""
    Text        : {token.text}
    POS         : {token.pos_}       # Part of speech
    Tag         : {token.tag_}       # Fine-grained POS tag
    Dep         : {token.dep_}       # Dependency role
    Head        : {token.head.text}  # Head word
    Lemma       : {token.lemma_}     # Base form
    Shape       : {token.shape_}     # Token shape (Xxxx, dddd, etc.)
    is_alpha    : {token.is_alpha}   # Is alphabetic?
    is_stop     : {token.is_stop}    # Is a stopword?
    """)

# Named Entity Recognition
print("\nNamed Entities:")
for ent in doc2.ents:
    print(f"{ent.text} → {ent.label_}")
```
